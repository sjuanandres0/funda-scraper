```python
#%%time
import scrapy
from scrapy.crawler import CrawlerProcess
from datetime import date
import re

class FundaSpider(scrapy.Spider):
    name = 'funda_spider_testing'
    allowed_domains = ["funda.nl"]
    download_delay = 2
    
    def start_requests(self, place='provincie-zuid-holland,provincie-noord-holland'):
#        places = ['leiden','hoofddorp','nieuw-vennep','voorhout','valkenburg-zh','alphen-aan-den-rijn',
#                  'sassenheim','hillegom','oegstgeest']
        places = ['leiden']
        places = ['heel-nederland']
        urls = ["http://www.funda.nl/koop/%s/p%s/" % (place, page_number) for place in places for page_number in range(20,100)]
        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse, meta={'dont_merge_cookies': True})

    def parse(self, response):
        links = response.css('div.search-result-media > a::attr(href)').extract()
        for link in links:
            yield response.follow(url=link, callback=self.parse2)
    
    def parse2(self, response):
        
        url = response.request.url        
        title = response.xpath('//title/text()').extract()[0]
        
        if re.search(r'/appartement-', url):
            property_type = 'apartment'
        elif re.search(r'/huis-', url):
            property_type = 'house'

        postal_code = re.search(r'\d{4} [A-Z]{2}', title).group(0)
        city = re.search(r'\d{4} [A-Z]{2} \w+',title).group(0).split()[2]
        address = re.findall(r'te koop: (.*) \d{4}',title)[0]
        
        try:
            price_kk = response.xpath('//*[@id="content"]/div/div/div[1]/section[4]/div/dl[1]/dd[1]/span[1]/text()').extract()
            price_kk = price_kk[0].replace('.','')
            price_kk = re.findall('[0-9]+', price_kk)[0]
        except:
            price_kk = 'n/a'
        
        try:
            price_m2 = response.xpath("//dt[contains(.,'Vraagprijs')]/following-sibling::dd[1]/text()").extract()[3]
            price_m2 = re.findall(r' \d+.\d+', price_m2)[0].strip().replace('.','')
        except:
            price_m2 = 'n/a'
            
        try:
            year_built = response.xpath('//*[@id="content"]/div/div/div[1]/section[4]/div/dl[2]/dd[3]/span[1]/text()').extract()[0]
        except:
            year_built = 'n/a'
            
        area_in = response.xpath('//*[@id="content"]/div/div/div[1]/section[4]/div/dl[3]/dd[2]/dl/dd[1]/span/text()').extract()[0]
        area_in = re.findall(r'\d+', area_in)[0]
        area_out = response.xpath('//*[@id="content"]/div/div/div[1]/section[4]/div/dl[3]/dd[3]/span/text()').extract()[0]
        area_out = re.findall(r'\d+', area_out)[0]

        try:
            rooms_base = response.xpath('//*[@id="content"]/div/div/div[1]/section[4]/div/dl[4]/dd[1]/span/text()').extract()[0]
            rooms = re.findall('\d+ kamer', rooms_base)[0].replace(' kamer','')
            bedrooms = re.findall('\d+ slaapkamer', rooms_base)[0].replace(' slaapkamer','')
        except:
            rooms = 'n/a'
            bedrooms = 'n/a'
        
        try:
            energie = response.xpath('//*[@id="content"]/div/div/div[1]/section[4]/div/dl[5]/dd[1]/span/@title').extract()[0]
            energie = energie.split()[1]
        except:
            energie = 'n/a'
        
        try:
            published = response.xpath('//*[@id="content"]/div/div/div[1]/section[4]/div/dl[1]/dd[3]/span/text()').extract()[0]
            published_2 = published
            if 'per maand' in published:
                published = response.xpath('//*[@id="content"]/div/div/div[1]/section[4]/div/dl[1]/dd[4]/span/text()').extract()[0]
                published_2 = published
            if published == 'Vandaag':
                published_2 = date.today().strftime("%d %B %Y")
        except:
            published = 'n/a'

        item = {'property_type':property_type, 'address':address, 'postal_code':postal_code, 'city':city, 'price_kk':price_kk,
                'price_m2':price_m2, 'year_built':year_built, 'area_in':area_in, 'area_out':area_out, 'rooms':rooms,
                'bedrooms':bedrooms, 'energie':energie, 'published':published, 'published_2':published_2, 'url':url}
        
        #funda_dict[url] = item
        funda_list.append(item)
        
#funda_dict = dict()
funda_list = []
                
process = CrawlerProcess()
process.crawl(FundaSpider)
process.start()

funda_list
```

```python
''' In lieu of geolocation, coordinates are looked up from the four numbers of the Dutch postal code only.'''
import pandas as pd
import numpy as np

df = pd.read_csv('funda_df_3.csv')

# Coordinates of 4-number Dutch postal codes, obtained from https://github.com/bobdenotter/4pp
df_4pp = pd.read_csv('4pp2021.csv')

def get_4pp(postal_code):
    return postal_code.split()[0]

def get_latitude(postal_code_4pp):
    try:
        return df_4pp[df_4pp['postcode']==int(postal_code_4pp)].iloc[0]['latitude']
    except:
        return np.nan

def get_longitude(postal_code_4pp):
    try:
        return df_4pp[df_4pp['postcode']==int(postal_code_4pp)].iloc[0]['longitude']
    except:
        return np.nan

df['postal_code_4pp'] = df['postal_code'].apply(get_4pp)
df['latitude_4pp'] = df['postal_code_4pp'].apply(get_latitude)
df['longitude_4pp'] = df['postal_code_4pp'].apply(get_longitude)

df
```
